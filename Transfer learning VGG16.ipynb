```python
import tensorflow
from tensorflow.keras.applications import VGG16
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg16 import preprocess_input,decode_predictions

```

```python
model=VGG16(weights="imagenet",include_top=True,input_shape=(224,224,3))
```



1.   VGG16 pretrained model 13 convolutional layer and 3 dense layer.
2.  
IMAGENET contains 1000 classes and the VGG16 allows the input shape of 224 x 224 with color image
(224 x 224 x 3).



```python
image1="/content/cat-6748193_1920.jpg"
```

```python
import cv2
cv=cv2.imread(image1)
cv.shape
```

```python
import matplotlib.image as mp
import matplotlib.pyplot as plt
```

```python
view_image=mp.imread(image1)
plt.imshow(view_image)
```

```python
new_image=cv2.resize(cv,(224,224))
new_image
```

```python
import numpy as np
from tensorflow.keras.preprocessing.image import img_to_array
```

```python
new_image=img_to_array(new_image)
new_image.shape
```

```python
x=new_image.reshape(1,224,224,3)
x.shape
```

```python
x
```

```python
x=preprocess_input(x)
x
```

```python
y=model.predict(x)
y
```

```python
decode_preds=decode_predictions(y,top=5)[0]
```

```python
for o in decode_preds:
  print(o[1],"->",o[2])
```
